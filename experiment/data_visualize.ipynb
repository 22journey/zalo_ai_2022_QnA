```python
wiki_data_file = "../data/wikipedia_20220620_cleaned/wikipedia_20220620_cleaned.jsonl"
wiki_titles_field = "../data/wikipedia_20220620_cleaned/wikipedia_20220620_all_titles.jsonl"

def read_data():
    with open(wiki_data_file, "r") as f:
        for line in f:
            yield line
```

```python
# load stopwords
stopwords_filename = [
    "../core/data/stop_words.txt",
    "../core/data/stop_words_2.txt",
]
stopwords = []
for sw in stopwords_filename:
    with open(sw, "r") as f:
        stopwords += [l.strip() for l in f]
stopwords = set(stopwords)
# print(stopwords)
```

```python
import json
from underthesea import text_normalize, word_tokenize
import string

SCORE_HIGHLIGHT = 3
SCORE_NORMAL = 1

def token_and_score(text):
    # normalize
    text = text_normalize(text)
    # remove unused link
    idx = text.find("BULLET ::::")
    # print(idx)
    if idx > 0:
        text = text[:idx]
    text = text.translate(str.maketrans('', '', string.punctuation))
    text = text.translate(str.maketrans('', '', string.digits))
    text = word_tokenize(text)
    d = {}
    for t in text:
        s = SCORE_HIGHLIGHT if t[0].isupper() else SCORE_NORMAL
        tl = t.lower()
        if tl not in stopwords:
            d[tl] = s 
    print(d)

import time
t = time.time()
iter = read_data()
for i in range(10):
    l = next(iter)
    if i < 1:
        continue
    text = json.loads(l)["text"]
    token_and_score(text)
    # print("\n")
    # if i % 100 == 0:
    #     print(i)
    #     print(time.time() - t)
```
